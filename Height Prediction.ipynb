# Linear Regression using Sklearn 

```python
# Importing various useful libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 

# Loading data into Python envrionmnet 
data = np.loadtxt('Pearson.txt')
# Getting height of only Father
father_height = data[:,0]
father_height = father_height.reshape(father_height.shape[0], 1)
# Getting height of only Son
son_height = data[:,1]
son_height = son_height.reshape(son_height.shape[0], 1)

# Div is int value of creating division in the data
div = round(0.7*father_height.shape[0]) # div = 70% of data in father_height

# Dividing father height data into test and train set
father_height_train = father_height[ :div, :]
father_height_test = father_height[ div + 1:, :]

# Dividing Son height data into test and train set
son_height_train = son_height[ :div, :]
son_height_test = son_height[ div + 1:, :]

# Training a Linear Regression model
model = linear_model.LinearRegression()
model.fit(father_height_train, son_height_train)

# Prediction results on test data
prediction = model.predict(father_height_test)

# Calculating error on the test data
error = (son_height_test - prediction) / prediction
percent_error = error*100
mean_error = percent_error.mean()
print('The mean error on the train data set is ', mean_error)

# Getting Learned parameters from the trained Linear Regression model
m = model.coef_
b = model.intercept_
y = m*father_height + b

# Plotting the data and result
plt.plot(father_height, y, 'r')
plt.scatter(father_height, son_height, s = 8)
plt.xlabel('Father Height')
plt.ylabel('Son Height ')
plt.title('Linear Regression')
```
