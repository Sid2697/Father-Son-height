```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('Pearson.txt')
father_height = data[:,0] #X
son_height = data[:,1] #Y
plt.scatter(father_height, son_height, s = 9)
plt.xlabel("Father Height")
plt.ylabel("Son Height")
plt.title('Father_height VS. Son_height')
```

```python
def gradientDescent(x, y, theta, alpha, num_iters):
    '''Function for calculating new values of hypothesis'''
    i = 0
#     theta = np.squeeze(np.asarray(theta))
    while i < num_iters:
        m = len(x) #No. of training examples
        h = theta.dot(x.T)
        theta[0,0] = theta[0,0] - (alpha/m)*sum(h - y)
        theta[1:,:] = theta[1:,:] - (alpha/m)*sum((h-y)*x)
        i += 1
    return theta
```

```python
initial_theta = np.array([[0],[0]])
alpha = 1
num_iters = 1000
theta = gradientDescent(father_height, son_height, initial_theta, alpha, num_iters)
```
